**Import Library**

```python
import pandas as pd
```

```python
import numpy as np
```

**Import CSV as DataFrame**

```python
df=pd.read_csv(r'https://github.com/YBI-Foundation/Dataset/raw/main/Boston.csv')
```

**Get the First Five Rows of Dataframe**


```python
df.head()
```

**Get Information of DataFrame**

```python
df.info()
```

**Get the Summary Statistics**

```python
df.describe()
```

**Get Shape of DataFrame**

```python
df.shape
```

**Define y (dependent or label or target variable) and X(indendent or features or attribute Variable)**

```python
y = df['MEDV']
```

```python
y.shape
```

```python
y
```

```python
x = df[['CRIM','ZN','INDUS','CHAS','NX','RM','AGE','DIS','RAD','TAX','PTRATIO','B','LSTAT']]
```

```python
x.shape
```

```python
x
```

**Get X variable Standardized**

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
mm = MinMaxScaler()
```

```python
x = mm.fit_transform(x)
```

```python
x
```

**Get Model Split**

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split (x,y,test_size = 0.3,random_state=2529)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

**Get Model Train**

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

**Get Model Prediction**

```python
y_pred = lr.predict(x_test)
```

```python
y_pred.shape
```

```python
y_pred
```

**Get Model Evaluation**

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python
mean_squared_error(y_test,y_pred)
```

```python
mean_absolute_error(y_test,y_pred)
```

```python
r2_score(y_test,y_pred)
```

**Get Visualization of Actual vs Predicted Results**

```python
import matplotlib.pyplot as plt
plt.scatter(y_test,y_pred)
plt.xlabel("Actual Prices")
plt.ylabel("Predicated prices")
plt.title("Actual price vs preicted price")
plt.show()
```
